```python
# import packages for web scraping
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = 'https://lemelson.mit.edu/resources/inventor-archive?page=0'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# view the html code
print(soup.prettify())

# find the inventor names
inventor_names = soup.find_all('h3', class_='field-content')
```

```python
inventor_names = soup.find_all('h4', class_='faux-full-title')
inventor_names = [name.text for name in inventor_names]
```

```python
#extract all entry after "<div class="field field--name-field-invention-name field--type-string field--label-hidden field__item">"
invention_names = soup.find_all('div', class_='field field--name-field-invention-name field--type-string field--label-hidden field__item')
invention_names = [name.text for name in invention_names]
```

```python
for i in range(31):
    url = 'https://lemelson.mit.edu/resources/inventor-archive?page=' + str(i)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    invention_names = soup.find_all('div', class_='field field--name-field-invention-name field--type-string field--label-hidden field__item')
    invention_names = [name.text for name in invention_names]
    if i == 0:
        df = pd.DataFrame({'Invention Name': invention_names})
    else:
        df = pd.concat([df, pd.DataFrame({'Invention Name': invention_names})])
    print(i)
```

```python
df.to_csv('lemelson_mit_inventor_archive.csv', index=False)
```
